```python
imge_rgb=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
```

```python
imge_gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
cv2.imshow('photos',imge_gray)
cv2.waitKey(0)
```

```python
imge_HSV=cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
```

```python
cv2.imshow('photos',imge_HSV)
cv2.waitKey(0)
```

```python
hue, sat,val=cv2.split(imge_HSV)
```

```python
val
```

```python
add_val=50
```

```python
import numpy as np
```

```python
new_val=val.astype(np.int64)+add_val
new_val[new_val>255]=255
new_val_uint8=new_val.astype(np.uint8)
```

```python
image_new_hsv=cv2.merge([hue,sat,new_val_uint8])
```

```python
image_bgr=cv2.cvtColor(image_new_hsv,cv2.COLOR_HSV2BGR)
```

```python
image_bgr
```

```python
cv2.imshow('photos',image_bgr)
cv2.waitKey(0)
```
